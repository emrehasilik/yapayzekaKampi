# Derin Öğrenmeye Giriş

Derin öğrenme, yapay zeka ve makine öğrenmesi alanlarında kullanılan bir alt dalı ifade eder. Bu teknik, bilgisayarların büyük miktarda veriyi işleyerek karmaşık problemleri çözmelerini sağlar. Derin öğrenme, sinir ağlarının çok katmanlı yapılarını kullanarak veri üzerinde otomatik olarak öğrenme ve temsil oluşturma yeteneğine dayanır.

Derin öğrenme, verinin hiyerarşik özelliklerini çıkarmak için katmanlı bir yaklaşım kullanır. Girdi verisi, katmanlar halinde işlenir ve her katman, önceki katmanın çıktısını temel alarak daha karmaşık özellikleri temsil eder. Bu sayede, daha karmaşık ve soyut kavramları öğrenmek mümkün hale gelir.

Derin öğrenme yöntemleri arasında evrişimli sinir ağları (CNN), rekurrent sinir ağları (RNN) ve derin yönlendirilmiş öğrenme (DRL) gibi teknikler bulunur. Bu teknikler, görüntü işleme, doğal dil işleme, ses tanıma, oyun oynama gibi çeşitli uygulamalarda kullanılır ve son yıllarda birçok alanda büyük başarılar elde edilmesini sağlamıştır.

## Makine Öğrenmesi & Derin Öğrenme

Makine öğrenmesi ve derin öğrenme, yapay zeka alanında kullanılan iki farklı teknikdir. İkisi de bilgisayarların veri üzerinde öğrenme yeteneğine dayanır, ancak farklı yaklaşımları vardır.

Makine Öğrenmesi:
Makine öğrenmesi, bir bilgisayarın veri üzerinde öğrenme yapabilmesi için kullanılan genel bir terimdir.

Makine öğrenmesi algoritmaları, belirli bir görevi gerçekleştirmek için veriden kalıpları tanımlar ve bu kalıpları kullanarak tahminler yapar veya kararlar alır.

Makine öğrenmesi algoritmaları, genellikle veri setleri üzerinde eğitilir ve bu eğitim sürecinde özellikler ve ilişkiler öğrenilir.

Makine öğrenmesi yöntemleri arasında destek vektör makineleri (SVM), karar ağaçları, rastgele ormanlar, k-NN (en yakın komşu) ve doğrusal regresyon gibi yöntemler bulunur.

Derin Öğrenme:
Derin öğrenme, makine öğrenmesinin bir alt dalıdır ve genellikle çok katmanlı yapılara dayanan sinir ağları kullanılarak gerçekleştirilir.

Derin öğrenme, genellikle büyük miktarda veri ve hesaplama gücü gerektirir, bu nedenle genellikle yüksek performanslı bilgisayarlar veya özel donanımlar kullanılır.

Derin öğrenme, veri üzerinde hiyerarşik özelliklerin otomatik olarak öğrenilmesini sağlar. Bu özellikler, girdi verisinden başlayarak her katmanda daha karmaşık ve soyut hale gelir.

Derin öğrenme, özellikle büyük ölçekli veri setleriyle başarılı olabilir ve genellikle görüntü işleme, doğal dil işleme, ses tanıma gibi alanlarda kullanılır.

Kısacası, derin öğrenme, makine öğrenmesinin bir alt kümesidir ve genellikle daha büyük ve daha karmaşık veri setlerinde daha etkili olabilir, ancak daha fazla hesaplama gücü gerektirir.

## Gözetimli Öğrenme & Gözetimsiz Öğrenme

Derin öğrenmede, gözetimli öğrenme ve gözetimsiz öğrenme olmak üzere iki ana yaklaşım bulunmaktadır.

**Gözetimli Öğrenme:**

* Gözetimli öğrenme, bir eğitim sürecinde makineye verilen girdi-veri çiftlerine dayanır. Bu çiftlerde, giriş verisiyle hedef çıktılar arasındaki ilişki açıkça belirtilir.

* Bu yaklaşım, öğrenme algoritmasının girdi verilerini inceleyerek bu veriler arasındaki kalıpları tanımasını ve hedef çıktıları tahmin etmesini sağlar.

* Örneğin, bir görüntü sınıflandırma görevinde, gözetimli öğrenme algoritması, etiketlenmiş bir veri setinde görüntülerle bunların sınıflarını eşleştirir ve daha sonra yeni görüntülerin sınıflarını tahmin edebilir.

**Gözetimsiz Öğrenme:**

* Gözetimsiz öğrenme, etiketlenmemiş veri kümeleri üzerinde çalışır. Bu yaklaşımda, verideki yapıları veya desenleri belirleyerek veri hakkında anlamlı bilgiler elde etmeye çalışır.

* Gözetimsiz öğrenme, veri setindeki gizli yapıları ve desenleri otomatik olarak ortaya çıkarır ve bu sayede verinin daha derinlemesine anlaşılmasını sağlar.

* Örneğin, kümeleme algoritmaları, gözetimsiz öğrenme örneklerine bir örnektir. Bu algoritmalar, veri setindeki benzer örnekleri gruplayarak verideki yapıları ortaya çıkarır.

Gözetimli öğrenme, genellikle belirli çıktılarla eğitilmiş veri setlerine dayanırken, gözetimsiz öğrenme, verideki yapıları bulmak ve anlamak için kullanılır. Gözetimli öğrenme daha yaygın olarak kullanılan bir yaklaşım olsa da, gözetimsiz öğrenme, belirli veri setlerindeki yapıları keşfetmek ve anlamak için çok değerlidir. Ayrıca, bu iki yaklaşımın bazen bir arada kullanıldığı yarı gözetimli veya takviyeli öğrenme gibi karmaşık teknikler de bulunmaktadır.

## Yapay Sinir Ağları

Yapay sinir ağları (YSN), biyolojik sinir ağlarından esinlenerek tasarlanmış matematiksel model ve algoritmaları içeren bir yapay zeka alanıdır. Bu sinir ağları, birçok bilgisayar görevini gerçekleştirebilen, öğrenme ve genelleştirme yetenekleri olan esnek bir yapıya sahiptir.

Yapay sinir ağları, birçok nöron adı verilen basit hesaplama birimlerinden oluşur. Her nöron, girdi verilerini alır, bu verileri işler ve bir çıkış üretir. Bu çıkışlar daha sonra diğer nöronlara veya çıkış katmanına iletilir. Nöronlar arasındaki bağlantılar, ağırlıklar adı verilen parametrelerle belirlenir ve bu ağırlıklar, ağın öğrenme sürecinde ayarlanır.

Yapay sinir ağlarının temel bileşenleri şunlardır:

1. Girdi Katmanı: Verinin sisteme giriş yaptığı katmandır. Bu katman, genellikle verinin her bir özelliğine karşılık gelen nöronlardan oluşur.

2. Gizli Katmanlar: Girdi katmanı ile çıkış katmanı arasında yer alan ve genellikle ağın karmaşıklığını temsil eden katmanlardır. Bir veya daha fazla gizli katman olabilir.

3. Çıkış Katmanı: Ağın sonucunu üreten katmandır. Genellikle bir sınıflandırma veya regresyon sonucu olarak yorumlanır.

Her nöron, girdileri ağırlıklarla çarparak bir aktivasyon fonksiyonuna (örneğin, sigmoid, ReLU gibi) sokarak çıktı üretir. Bu aktivasyon fonksiyonu, nöronun çıktısını belirler ve belirli bir eşiğin üzerinde olup olmadığını kontrol eder.

Yapay sinir ağları, öğrenme sürecinde genellikle geriye yayılım (backpropagation) algoritması kullanılarak eğitilir. Bu süreçte, ağın tahminleri gerçek etiketlerle karşılaştırılır ve hata (kayıp) hesaplanır. Ardından, bu hata geriye doğru yayılarak ağın ağırlıkları güncellenir, böylece hatayı minimize eden bir model elde edilir.

Yapay sinir ağları, görüntü tanıma, doğal dil işleme, ses tanıma, sınıflandırma, regresyon ve çeşitli diğer görevlerde geniş bir uygulama yelpazesine sahiptir.

## Aktivasyon Fonksiyonları

Aktivasyon fonksiyonları, yapay sinir ağlarında kullanılan bir tür matematiksel işlemdir. Bu fonksiyonlar, sinir ağlarının katmanlar arasındaki çıktıları şekillendirmek ve modelin doğrusallık dışındaki karmaşıklıkları öğrenmesine yardımcı olmak için kullanılır.

Aktivasyon fonksiyonları, sinir ağlarının esnekliğini ve öğrenme yeteneğini etkileyebilir. Dolayısıyla, belirli bir probleme en uygun aktivasyon fonksiyonunun seçilmesi önemlidir.

### Sigmoid Fonksiyonu

Sigmoid fonksiyonu, genellikle girişleri 0 ile 1 arasında sınırlar. Bu, sınırlı çıkış aralığı, bir nöronun aktivasyonunu hesaplarken kullanışlı olabilir. Ancak, büyük derecelerde girişlerde türevi çok küçük olduğundan, derin sinir ağlarında kullanımı yaygın değildir.

### ReLU (Rectified Linear Unit) Fonksiyonu

ReLU fonksiyonu, negatif girişleri sıfıra eşitlerken pozitif girişlerin aynı kalmasını sağlar. Bu fonksiyon, son zamanlarda sinir ağlarında çok popüler hale gelmiştir çünkü hesaplama açısından ucuzdur ve öğrenme açısından da iyi sonuçlar verir.

ReLU-activation-function.png

### Leaky ReLU (Sızan Lineer Birim) Fonksiyonu

ReLU'nun genel bir sorunu, negatif girişler için sıfır değeri döndürmesidir. Bu, bazı durumlarda "ölü nöronlar" problemine neden olabilir. Leaky ReLU, negatif girişler için küçük bir eğim ile düzeltilmiş bir versiyondur.

### Tanh Fonksiyonu

Tanh fonksiyonu, sigmoid fonksiyonuna benzer ancak çıktı aralığı -1 ile 1 arasındadır. Bu nedenle, sıfır merkezli olması ve sigmoid'e göre daha simetrik bir şekilde dağıtılması nedeniyle bazı durumlarda tercih edilir.

Basliksiz.png

### Softmax Fonksiyonu

Softmax fonksiyonu, bir sinir ağının çoklu sınıf sınıflandırma problemlerinde kullanılan bir aktivasyon fonksiyonudur. Çıkışları bir olasılık dağılımı haline getirir, yani tüm çıkışların toplamı 1'e eşittir, böylece sınıf olasılıklarını temsil eder.

## İleri Yayılım ve Geri Yayılım


İleri yayılım (forward propagation) ve geri yayılım (backpropagation), yapay sinir ağlarının eğitimi sırasında kullanılan iki temel adımdır.

İleri Yayılım (Forward Propagation):

* İleri yayılım, sinir ağına bir giriş verisi uygulandığında, bu verinin ağın her katmanından geçerek sonuç çıktısının hesaplanması sürecidir.
* Giriş verisi, girdi katmanından başlayarak sinir ağının her bir katmanına doğru aktarılır. Her katmanda, giriş verisi bir dizi matematiksel işleme tabi tutulur ve bir çıktı üretilir.
* İleri yayılım sürecinde, her bir katmanın çıktısı bir sonraki katmanın girdisi olur ve bu işlem, ağın son katmanına kadar devam eder. Son katmanda, ağın tahmin ettiği çıktı elde edilir.

Geri Yayılım (Backpropagation):
* Geri yayılım, sinir ağının eğitimi sırasında, tahmin edilen çıktı ile gerçek çıktı arasındaki hata miktarının hesaplanması ve bu hatanın ağın geriye doğru her bir katmanına yayılması sürecidir.
* Hatanın geriye doğru yayılması, her bir katmandaki ağırlık ve bias değerlerini güncellemek için kullanılır. Bu sayede, ağın çıktısı üzerindeki hata azaltılarak daha iyi tahminler elde edilir.
* Geri yayılım algoritması, zincir kuralı ve gradyan iniş algoritmasıyla birlikte kullanılır. Hata, zincir kuralı kullanılarak her katmanda hesaplanır ve ardından gradyan iniş algoritması kullanılarak ağırlık ve bias değerleri güncellenir.
* Geri yayılım süreci, genellikle bir eğitim verisi setindeki tüm örnekler üzerinde tekrarlanarak gerçekleştirilir. Bu sayede, ağın tüm parametreleri optimize edilir ve istenilen sonuçlar elde edilir.

İleri yayılım ve geri yayılım, sinir ağlarının öğrenme sürecinde birlikte çalışarak ağın doğru şekilde eğitilmesini sağlar. İleri yayılım, girdi verisinden çıktıyı hesaplamak için kullanılırken, geri yayılım, hesaplanan hatayı kullanarak ağın parametrelerini güncellemek için kullanılır. Bu süreç, sinir ağlarının çeşitli görevleri öğrenmesini ve karmaşık problemleri çözmesini sağlar.

### Optimizasyon

Optimizasyon, bir hedefi en iyi şekilde karşılayan bir çözüm bulma sürecidir.

Yapay sinir ağlarında optimizasyon, genellikle bir kayıp fonksiyonunu minimize ederek (veya maksimize ederek, örneğin doğruluk) ağın performansını iyileştirmeyi amaçlar.

Bu optimizasyon süreci, ağdaki ağırlıkların ve önyargıların güncellenmesini içerir. Bunun için genellikle gradyan iniş gibi optimizasyon algoritmaları kullanılır.

Amaç, eğitim verileri üzerinde iyi bir performans elde etmek ve ağın genelleme yapabilmesini sağlamaktır.

İleri yayılım, sinir ağının bir girdiyi alıp bir tahmin üretmesini sağlar. Geri yayılım, bu tahminin gerçek değerlerle karşılaştırılması ve ağdaki hataların geriye doğru iletilmesi için kullanılır. Son olarak, optimizasyon süreci, ağırlıkların ve önyargıların güncellenmesiyle ağın performansını iyileştirir.

Yaygın olarak kullanılan bazı optimizasyon algoritmaları:

**Stokastik Gradyan Azalma (SGD):**

* Bu, en temel optimizasyon algoritmalarından biridir. Her eğitim örneği için ağırlıkları güncellemek için gradyan hesaplanır.

* Ancak, bu yaklaşım, gradyanların büyük varyanslara sahip olduğu durumlarda dalgalanmaya neden olabilir ve minimuma ulaşmak için zaman alabilir.

* Bu nedenle, daha gelişmiş optimizasyon teknikleri, SGD'nin bu zayıf yönlerini ele alır.

**Adam (Adaptive Moment Estimation):**

* Adam, stokastik gradyan azalma algoritmasının geliştirilmiş bir versiyonudur. Hem gradyanların hem de gradyanların karesinin hareketli ortalamalarını hesaplar.

* Bu, gradyanların büyük varyansları ile başa çıkmaya ve farklı öğrenme oranlarını adaptif olarak ayarlamaya yardımcı olur.

* Adam, genellikle derin öğrenme modellerinde etkili bir şekilde kullanılır.

**Adadelta:**

* Adadelta, gradyan azalma algoritmasının bir türüdür ve özellikle öğrenme oranını ayarlamak için adaptif bir mekanizma içerir.

* Adadelta, önceki gradyanların ikinci momentlerini izler ve öğrenme oranını gradyanlara göre ayarlar.

* Bu, öğrenme oranı hiperparametresine ihtiyaç duymadan gradyanlara göre otomatik olarak uyum sağlar.

Bu optimizasyon algoritmaları, modelin eğitim sürecinde kullanılan ve farklı problemler ve veri setleri için farklı performanslar sergileyebilen yaygın olarak kullanılan örneklerdir.
